## Part 1: Quadratic Forms and Conversions

### ðŸŽ¯ Objectives
- Create quadratic functions in different forms
- Convert between standard, vertex, and factored forms
- Find vertex, axis of symmetry, and intercepts


````python
class QuadraticFunction:
    """
    Represents a quadratic function with multiple forms and analysis methods.
    """
    
    def __init__(self, a=None, b=None, c=None, h=None, k=None, r1=None, r2=None):
        """
        Create quadratic from various forms.
        
        Parameters:
        -----------
        a, b, c : Standard form f(x) = axÂ² + bx + c
        a, h, k : Vertex form f(x) = a(x-h)Â² + k
        a, r1, r2 : Factored form f(x) = a(x-r1)(x-r2)
        """
        if a is not None and b is not None and c is not None:
            # Standard form
            self.a, self.b, self.c = a, b, c
            self.h = -b / (2*a)
            self.k = a * self.h**2 + b * self.h + c
        elif a is not None and h is not None and k is not None:
            # Vertex form
            self.a, self.h, self.k = a, h, k
            self.b = -2 * a * h
            self.c = a * h**2 + k
        elif a is not None and r1 is not None and r2 is not None:
            # Factored form
            self.a = a
            self.b = -a * (r1 + r2)
            self.c = a * r1 * r2
            self.h = -self.b / (2*self.a)
            self.k = self.evaluate(self.h)
        else:
            raise ValueError("Invalid parameters for quadratic function")
    
    def evaluate(self, x):
        """Evaluate f(x)."""
        return self.a * x**2 + self.b * x + self.c
    
    def standard_form(self):
        """Return standard form string."""
        b_sign = '+' if self.b >= 0 else ''
        c_sign = '+' if self.c >= 0 else ''
        return f"f(x) = {self.a}xÂ² {b_sign}{self.b}x {c_sign}{self.c}"
    
    def vertex_form(self):
        """Return vertex form string."""
        h_sign = '-' if self.h >= 0 else '+'
        k_sign = '+' if self.k >= 0 else ''
        return f"f(x) = {self.a}(x {h_sign}{abs(self.h)})Â² {k_sign}{self.k}"
    
    def find_roots(self):
        """Find x-intercepts using quadratic formula."""
        discriminant = self.b**2 - 4*self.a*self.c
        
        if discriminant > 0:
            root1 = (-self.b + np.sqrt(discriminant)) / (2*self.a)
            root2 = (-self.b - np.sqrt(discriminant)) / (2*self.a)
            return (root1, root2), "two distinct real roots"
        elif discriminant == 0:
            root = -self.b / (2*self.a)
            return (root,), "one repeated real root"
        else:
            real_part = -self.b / (2*self.a)
            imag_part = np.sqrt(abs(discriminant)) / (2*self.a)
            return (complex(real_part, imag_part), complex(real_part, -imag_part)), "two complex roots"
    
    def get_discriminant(self):
        """Calculate discriminant."""
        return self.b**2 - 4*self.a*self.c
    
    def get_vertex(self):
        """Return vertex coordinates."""
        return (self.h, self.k)
    
    def get_y_intercept(self):
        """Return y-intercept."""
        return self.c
    
    def get_axis_of_symmetry(self):
        """Return axis of symmetry."""
        return self.h
    
    def opens_upward(self):
        """Check if parabola opens upward."""
        return self.a > 0
    
    def __repr__(self):
        return self.standard_form()

# Test with examples
print("="*70)
print("QUADRATIC FUNCTIONS: FORMS AND CONVERSIONS")
print("="*70)

# Example 1: Standard form
print("\n1. Standard Form: f(x) = 2xÂ² - 8x + 6")
q1 = QuadraticFunction(a=2, b=-8, c=6)
print(f"   Standard: {q1.standard_form()}")
print(f"   Vertex: {q1.vertex_form()}")
print(f"   Vertex coordinates: {q1.get_vertex()}")
print(f"   Y-intercept: {q1.get_y_intercept()}")
print(f"   Axis of symmetry: x = {q1.get_axis_of_symmetry()}")
print(f"   Opens: {'upward â˜º' if q1.opens_upward() else 'downward â˜¹'}")
roots, root_type = q1.find_roots()
print(f"   Roots: {roots} ({root_type})")
print(f"   Discriminant: {q1.get_discriminant()}")

# Example 2: Vertex form
print("\n2. Vertex Form: f(x) = -1(x - 3)Â² + 4")
q2 = QuadraticFunction(a=-1, h=3, k=4)
print(f"   Standard: {q2.standard_form()}")
print(f"   Vertex: {q2.vertex_form()}")
print(f"   Vertex coordinates: {q2.get_vertex()}")
print(f"   Opens: {'upward â˜º' if q2.opens_upward() else 'downward â˜¹'}")
roots, root_type = q2.find_roots()
print(f"   Roots: {roots} ({root_type})")

# Example 3: Factored form
print("\n3. Factored Form: f(x) = (x - 1)(x - 5)")
q3 = QuadraticFunction(a=1, r1=1, r2=5)
print(f"   Standard: {q3.standard_form()}")
print(f"   Vertex: {q3.vertex_form()}")
print(f"   Vertex coordinates: {q3.get_vertex()}")
print(f"   Y-intercept: {q3.get_y_intercept()}")

# Example 4: No real roots
print("\n4. No Real Roots: f(x) = xÂ² + 2x + 5")
q4 = QuadraticFunction(a=1, b=2, c=5)
print(f"   Standard: {q4.standard_form()}")
print(f"   Vertex: {q4.vertex_form()}")
roots, root_type = q4.find_roots()
print(f"   Roots: {roots} ({root_type})")
print(f"   Discriminant: {q4.get_discriminant()} < 0 â†’ no real roots")
````

## Part 2: Completing the Square

### ðŸŽ¯ Objectives
- Master completing the square technique
- Convert standard form to vertex form algebraically
- Understand geometric interpretation


````python
def complete_the_square(a, b, c):
    """
    Complete the square for axÂ² + bx + c.
    
    Returns vertex form parameters (a, h, k).
    """
    # Factor out 'a' from first two terms
    # axÂ² + bx + c = a(xÂ² + (b/a)x) + c
    
    # Complete the square inside
    # xÂ² + (b/a)x = (x + b/(2a))Â² - (b/(2a))Â²
    
    h = -b / (2*a)
    k = c - (b**2) / (4*a)
    
    return a, h, k

# Examples of completing the square
print("="*70)
print("COMPLETING THE SQUARE")
print("="*70)

examples = [
    (1, 6, 5, "xÂ² + 6x + 5"),
    (2, -8, 6, "2xÂ² - 8x + 6"),
    (-1, 4, -3, "-xÂ² + 4x - 3"),
    (3, 12, 7, "3xÂ² + 12x + 7"),
]

for a_val, b_val, c_val, expr in examples:
    print(f"\n{expr}")
    print(f"Standard form: f(x) = {a_val}xÂ² + {b_val}x + {c_val}")
    
    # Complete the square
    a_new, h, k = complete_the_square(a_val, b_val, c_val)
    
    # Show steps
    print(f"\nStep 1: Factor out {a_val} from first two terms")
    print(f"        f(x) = {a_val}(xÂ² + {b_val/a_val}x) + {c_val}")
    
    print(f"\nStep 2: Complete the square inside")
    half_b_over_a = b_val / (2*a_val)
    print(f"        Half of {b_val/a_val} is {half_b_over_a}")
    print(f"        Square it: ({half_b_over_a})Â² = {half_b_over_a**2}")
    
    print(f"\nStep 3: Add and subtract ({half_b_over_a})Â² inside")
    print(f"        f(x) = {a_val}[(x + {half_b_over_a})Â² - {half_b_over_a**2}] + {c_val}")
    
    print(f"\nStep 4: Simplify")
    h_sign = '-' if h >= 0 else '+'
    k_sign = '+' if k >= 0 else ''
    print(f"        f(x) = {a_new}(x {h_sign}{abs(h)})Â² {k_sign}{k}")
    
    print(f"\nVertex: ({h}, {k})")
    print(f"{'Maximum' if a_val < 0 else 'Minimum'} value: {k}")
    
    # Verify
    q = QuadraticFunction(a=a_val, b=b_val, c=c_val)
    assert np.isclose(q.h, h) and np.isclose(q.k, k), "Verification failed!"
    print("âœ… Verified!")
    print("-" * 70)
````

## Part 3: Optimization Problems

### ðŸŽ¯ Objectives

````python
print("="*70)
print("OPTIMIZATION PROBLEMS")
print("="*70)

# Problem 1: Maximize Area with Fixed Perimeter
print("\nðŸ“ PROBLEM 1: Rectangular Enclosure")
print("A farmer has 200 meters of fencing to enclose a rectangular field.")
print("One side is along a river (no fence needed there).")
print("What dimensions maximize the enclosed area?")
print()

# Let width = w, then length = 200 - 2w (since two widths + one length = 200)
# Area A(w) = w Ã— (200 - 2w) = 200w - 2wÂ²

def area_function(w):
    return 200*w - 2*w**2

# This is a quadratic: A(w) = -2wÂ² + 200w + 0
a, b, c = -2, 200, 0
q_area = QuadraticFunction(a=a, b=b, c=c)

print(f"Area function: A(w) = {q_area.standard_form()}")
print(f"Vertex form: A(w) = {q_area.vertex_form()}")

w_opt, area_max = q_area.get_vertex()
length_opt = 200 - 2*w_opt

print(f"\nOptimal width: {w_opt} meters")
print(f"Optimal length: {length_opt} meters")
print(f"Maximum area: {area_max} square meters")

# Visualize
w_range = np.linspace(0, 100, 200)
area_values = area_function(w_range)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: Area vs Width
ax1.plot(w_range, area_values, 'b-', linewidth=3)
ax1.scatter([w_opt], [area_max], s=300, c='red', marker='*', edgecolors='black', linewidths=2, zorder=5)
ax1.axvline(x=w_opt, color='red', linestyle='--', alpha=0.5)
ax1.axhline(y=area_max, color='red', linestyle='--', alpha=0.5)
ax1.set_xlabel('Width (meters)', fontsize=12, fontweight='bold')
ax1.set_ylabel('Area (sq meters)', fontsize=12, fontweight='bold')
ax1.set_title(f'Maximize Area with Fixed Fencing\nOptimal: {w_opt}m Ã— {length_opt}m = {area_max} mÂ²', fontsize=12, fontweight='bold')
ax1.grid(True, alpha=0.3)
ax1.text(w_opt+5, area_max-200, f'Maximum\n({w_opt:.0f}, {area_max:.0f})', fontsize=10, bbox=dict(boxstyle='round', facecolor='yellow'))

# Plot 2: Rectangle visualization
ax2.set_xlim(-10, 110)
ax2.set_ylim(-10, 60)
rect = plt.Rectangle((0, 0), length_opt, w_opt, fill=False, edgecolor='blue', linewidth=3)
ax2.add_patch(rect)
ax2.plot([0, length_opt], [0, 0], 'cyan', linewidth=5, label='River (no fence)')
ax2.plot([0, 0], [0, w_opt], 'brown', linewidth=3, label='Fence')
ax2.plot([length_opt, length_opt], [0, w_opt], 'brown', linewidth=3)
ax2.plot([0, length_opt], [w_opt, w_opt], 'brown', linewidth=3)
ax2.text(length_opt/2, w_opt/2, f'Area = {area_max} mÂ²', ha='center', va='center', fontsize=14, fontweight='bold', bbox=dict(boxstyle='round', facecolor='lightgreen'))
ax2.text(length_opt/2, -5, f'{length_opt} m', ha='center', fontsize=11, fontweight='bold')
ax2.text(-5, w_opt/2, f'{w_opt} m', va='center', fontsize=11, fontweight='bold', rotation=90)
ax2.set_aspect('equal')
ax2.legend(loc='upper right')
ax2.set_title('Optimal Rectangle', fontsize=12, fontweight='bold')
ax2.axis('off')

plt.tight_layout()
plt.show()

print("\n" + "="*70)

# Problem 2: Minimize Cost
print("\nðŸ’° PROBLEM 2: Product Pricing")
print("A company's daily revenue is R(p) = -2pÂ² + 80p - 300")
print("where p is the price per unit (in dollars).")
print("Find the price that maximizes revenue.")
print()

a, b, c = -2, 80, -300
q_revenue = QuadraticFunction(a=a, b=b, c=c)

p_opt, revenue_max = q_revenue.get_vertex()

print(f"Revenue function: R(p) = {q_revenue.standard_form()}")
print(f"Vertex form: R(p) = {q_revenue.vertex_form()}")
print(f"\nOptimal price: ${p_opt:.2f}")
print(f"Maximum revenue: ${revenue_max:.2f}")

# Problem 3: Projectile Motion
print("\n" + "="*70)
print("\nðŸŽ¯ PROBLEM 3: Projectile Motion")
print("A ball is thrown upward with height h(t) = -5tÂ² + 20t + 2 meters")
print("where t is time in seconds.")
print()

a, b, c = -5, 20, 2
q_height = QuadraticFunction(a=a, b=b, c=c)

t_max, h_max = q_height.get_vertex()

print(f"Height function: h(t) = {q_height.standard_form()}")
print(f"Vertex form: h(t) = {q_height.vertex_form()}")
print(f"\nMaximum height: {h_max} meters")
print(f"Time at max height: {t_max} seconds")

# When does it hit the ground? h(t) = 0
roots, _ = q_height.find_roots()
t_ground = max(roots)  # Take the positive root
print(f"Hits ground at: {t_ground:.2f} seconds")

# Visualize
t_range = np.linspace(0, t_ground, 200)
height = q_height.evaluate(t_range)

plt.figure(figsize=(10, 6))
plt.plot(t_range, height, 'b-', linewidth=3, label='Trajectory')
plt.scatter([t_max], [h_max], s=300, c='red', marker='*', edgecolors='black', linewidths=2, label=f'Peak ({t_max}s, {h_max}m)', zorder=5)
plt.scatter([0], [2], s=150, c='green', marker='o', edgecolors='black', linewidths=2, label='Launch (0s, 2m)', zorder=5)
plt.scatter([t_ground], [0], s=150, c='orange', marker='X', edgecolors='black', linewidths=2, label=f'Landing ({t_ground:.2f}s, 0m)', zorder=5)
plt.axhline(y=0, color='brown', linewidth=2, label='Ground')
plt.xlabel('Time (seconds)', fontsize=12, fontweight='bold')
plt.ylabel('Height (meters)', fontsize=12, fontweight='bold')
plt.title('Projectile Motion: Ball Trajectory', fontsize=14, fontweight='bold')
plt.legend(loc='upper right', fontsize=10)
plt.grid(True, alpha=0.3)
plt.show()

print("ðŸ“Š Optimization visualizations complete!")
````

## Part 4: Polynomial Regression (ML Application)

### ðŸŽ¯ Objectives
- Fit quadratic models to data
- Compare linear vs quadratic regression
- Evaluate model performance with RÂ²
- Understand when to use polynomial features


````python
print("="*70)
print("POLYNOMIAL REGRESSION: LINEAR VS QUADRATIC")
print("="*70)

# Generate synthetic data with quadratic relationship
np.random.seed(42)
X_orig = np.linspace(0, 10, 50)
y_true = 0.5 * X_orig**2 - 3 * X_orig + 10  # True quadratic relationship
y_noisy = y_true + np.random.normal(0, 3, len(X_orig))  # Add noise

print("\nGenerated 50 data points with quadratic relationship:")
print("True function: y = 0.5xÂ² - 3x + 10 (plus noise)")
print()

# Reshape for sklearn
X_orig_reshaped = X_orig.reshape(-1, 1)

# Model 1: Linear Regression
print("MODEL 1: Linear Regression (Baseline)")
lr_linear = LinearRegression()
lr_linear.fit(X_orig_reshaped, y_noisy)
y_pred_linear = lr_linear.predict(X_orig_reshaped)
r2_linear = r2_score(y_noisy, y_pred_linear)
mse_linear = mean_squared_error(y_noisy, y_pred_linear)

print(f"Equation: y = {lr_linear.coef_[0]:.2f}x + {lr_linear.intercept_:.2f}")
print(f"RÂ² score: {r2_linear:.4f}")
print(f"MSE: {mse_linear:.2f}")

# Model 2: Quadratic Regression (Polynomial Features)
print("\nMODEL 2: Quadratic Regression (Polynomial Features)")
poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(X_orig_reshaped)
print(f"Features: {poly.get_feature_names_out(['x'])}")

lr_poly = LinearRegression()
lr_poly.fit(X_poly, y_noisy)
y_pred_poly = lr_poly.predict(X_poly)
r2_poly = r2_score(y_noisy, y_pred_poly)
mse_poly = mean_squared_error(y_noisy, y_pred_poly)

print(f"Equation: y = {lr_poly.coef_[1]:.2f}xÂ² + {lr_poly.coef_[0]:.2f}x + {lr_poly.intercept_:.2f}")
print(f"RÂ² score: {r2_poly:.4f}")
print(f"MSE: {mse_poly:.2f}")

print("\n" + "="*70)
print("COMPARISON")
print("="*70)
print(f"Linear RÂ²:    {r2_linear:.4f}")
print(f"Quadratic RÂ²: {r2_poly:.4f}")
print(f"Improvement:  {(r2_poly - r2_linear)*100:.2f}%")
print()
print(f"Linear MSE:    {mse_linear:.2f}")
print(f"Quadratic MSE: {mse_poly:.2f}")
print(f"Reduction:     {(1 - mse_poly/mse_linear)*100:.2f}%")

# Visualize
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Plot 1: Linear vs Quadratic
ax1 = axes[0]
ax1.scatter(X_orig, y_noisy, s=50, alpha=0.6, c='gray', edgecolors='black', label='Data')
ax1.plot(X_orig, y_true, 'g--', linewidth=2, alpha=0.7, label='True function (quadratic)')
ax1.plot(X_orig, y_pred_linear, 'b-', linewidth=3, label=f'Linear fit (RÂ²={r2_linear:.3f})')
ax1.plot(X_orig, y_pred_poly, 'r-', linewidth=3, label=f'Quadratic fit (RÂ²={r2_poly:.3f})')
ax1.set_xlabel('x', fontsize=12, fontweight='bold')
ax1.set_ylabel('y', fontsize=12, fontweight='bold')
ax1.set_title('Linear vs Quadratic Regression', fontsize=12, fontweight='bold')
ax1.legend(loc='upper right', fontsize=9)
ax1.grid(True, alpha=0.3)

# Plot 2: Residuals for Linear
ax2 = axes[1]
residuals_linear = y_noisy - y_pred_linear
ax2.scatter(X_orig, residuals_linear, s=50, c='blue', alpha=0.6, edgecolors='black')
ax2.axhline(y=0, color='red', linestyle='--', linewidth=2)
ax2.set_xlabel('x', fontsize=12, fontweight='bold')
ax2.set_ylabel('Residuals', fontsize=12, fontweight='bold')
ax2.set_title(f'Linear Model Residuals\nMSE = {mse_linear:.2f}', fontsize=12, fontweight='bold')
ax2.grid(True, alpha=0.3)

# Plot 3: Residuals for Quadratic
ax3 = axes[2]
residuals_poly = y_noisy - y_pred_poly
ax3.scatter(X_orig, residuals_poly, s=50, c='red', alpha=0.6, edgecolors='black')
ax3.axhline(y=0, color='red', linestyle='--', linewidth=2)
ax3.set_xlabel('x', fontsize=12, fontweight='bold')
ax3.set_ylabel('Residuals', fontsize=12, fontweight='bold')
ax3.set_title(f'Quadratic Model Residuals\nMSE = {mse_poly:.2f}', fontsize=12, fontweight='bold')
ax3.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\nðŸ“Š Polynomial regression comparison complete!")
print("\nðŸ’¡ Key Insight: Quadratic model fits curved data much better!")
print("   When scatter plot shows curvature, add xÂ² feature!")
````
