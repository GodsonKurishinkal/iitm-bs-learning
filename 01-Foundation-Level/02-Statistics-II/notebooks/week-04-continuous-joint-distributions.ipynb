## 1. Bivariate Normal Distribution
### 1.1 Theory
The bivariate Normal extends the Normal to two dimensions with correlation. It's fundamental for multivariate analysis.

### 1.2 Mathematical Definition
$(X, Y) \sim N(\mu_X, \mu_Y, \sigma_X^2, \sigma_Y^2, \rho)$

**Marginals**: $X \sim N(\mu_X, \sigma_X^2)$, $Y \sim N(\mu_Y, \sigma_Y^2)$

**Conditional**: $Y|X=x \sim N\left(\mu_Y + \rho\frac{\sigma_Y}{\sigma_X}(x - \mu_X), \sigma_Y^2(1 - \rho^2)\right)$
