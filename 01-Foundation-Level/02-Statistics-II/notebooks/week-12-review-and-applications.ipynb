## Formula Sheet

### Estimation
| Formula | Description |
|---------|------------|
| $\bar{X} \pm t_{\alpha/2} \frac{s}{\sqrt{n}}$ | CI for mean |
| $\hat{p} \pm z_{\alpha/2} \sqrt{\frac{\hat{p}(1-\hat{p})}{n}}$ | CI for proportion |

### Hypothesis Tests
| Test | Statistic |
|------|----------|
| One-sample t | $t = \frac{\bar{X} - \mu_0}{s/\sqrt{n}}$ |
| Two-sample t | $t = \frac{\bar{X}_1 - \bar{X}_2}{s_p\sqrt{1/n_1 + 1/n_2}}$ |
| Paired t | $t = \frac{\bar{d}}{s_d/\sqrt{n}}$ |

### Regression
| Formula | Description |
|---------|------------|
| $\hat{\beta}_1 = r\frac{s_Y}{s_X}$ | Slope |
| $R^2 = 1 - \frac{SS_{res}}{SS_{tot}}$ | Coefficient of determination |
