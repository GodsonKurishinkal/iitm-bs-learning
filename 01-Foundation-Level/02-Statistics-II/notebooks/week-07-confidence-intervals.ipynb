## 1. Confidence Intervals for Means
### 1.1 Theory
A confidence interval quantifies uncertainty in our point estimate. 95% CI means 95% of such intervals contain the true parameter.

### 1.2 Mathematical Definition
**Known variance**: $\bar{X} \pm z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}}$

**Unknown variance**: $\bar{X} \pm t_{\alpha/2, n-1} \cdot \frac{s}{\sqrt{n}}$

## 2. The t-Distribution
### 2.1 Theory
When σ is unknown and estimated from data, we use the t-distribution with heavier tails than Normal.

### 2.2 Mathematical Definition
$$T = \frac{\bar{X} - \mu}{S/\sqrt{n}} \sim t_{n-1}$$

As $n \to \infty$, $t \to N(0,1)$

### 3.1 Theory
For proportions (like defect rate or conversion rate), we use Normal approximation.

### 3.2 Mathematical Definition
$$\hat{p} \pm z_{\alpha/2} \sqrt{\frac{\hat{p}(1-\hat{p})}{n}}$$
