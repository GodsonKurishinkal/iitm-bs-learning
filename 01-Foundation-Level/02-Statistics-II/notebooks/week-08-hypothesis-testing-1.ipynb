## 1. Hypothesis Testing Framework
### 1.1 Theory
Hypothesis testing evaluates claims about population parameters using sample data.

### 1.2 Mathematical Definition
- $H_0$: Null hypothesis (status quo)
- $H_1$: Alternative hypothesis (claim to test)
- **p-value**: P(data as extreme or more | $H_0$ true)
- **Decision**: Reject $H_0$ if p-value < α

| Fail to Reject | Correct | Type II Error (β) |
