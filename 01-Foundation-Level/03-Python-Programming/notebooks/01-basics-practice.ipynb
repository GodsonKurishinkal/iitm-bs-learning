## 1. Variables and Data Types

### Learning Objectives:
- Understand different data types in Python
- Practice variable assignment
- Learn type conversion

```python
# Integer
age = 25
print(f"Age: {age}, Type: {type(age)}")

# Float
height = 5.8
print(f"Height: {height}, Type: {type(height)}")

# String
name = "Godson"
print(f"Name: {name}, Type: {type(name)}")

# Boolean
is_student = True
print(f"Is Student: {is_student}, Type: {type(is_student)}")
```

## 2. Basic Arithmetic Operations

```python
# Basic operations
a = 10
b = 3

print(f"Addition: {a} + {b} = {a + b}")
print(f"Subtraction: {a} - {b} = {a - b}")
print(f"Multiplication: {a} * {b} = {a * b}")
print(f"Division: {a} / {b} = {a / b}")
print(f"Floor Division: {a} // {b} = {a // b}")
print(f"Modulus: {a} % {b} = {a % b}")
print(f"Exponentiation: {a} ** {b} = {a ** b}")
```

## 3. Strings and String Operations

```python
# String operations
text = "IIT Madras Data Science"

print(f"Original: {text}")
print(f"Uppercase: {text.upper()}")
print(f"Lowercase: {text.lower()}")
print(f"Length: {len(text)}")
print(f"Split: {text.split()}")
print(f"Replace: {text.replace('Data Science', 'BS Degree')}")
```

## 4. Lists - Introduction

```python
# Creating lists
numbers = [1, 2, 3, 4, 5]
subjects = ["Math", "Statistics", "Python", "CT", "English"]
mixed = [1, "Hello", 3.14, True]

print(f"Numbers: {numbers}")
print(f"Subjects: {subjects}")
print(f"Mixed: {mixed}")

# List operations
print(f"\nFirst subject: {subjects[0]}")
print(f"Last subject: {subjects[-1]}")
print(f"First three: {subjects[:3]}")
```

## 5. Conditional Statements

```python
# If-else example
score = 85

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"Score: {score}, Grade: {grade}")
```

## 6. Loops

```python
# For loop
print("For loop example:")
for i in range(1, 6):
    print(f"Number: {i}")

# While loop
print("\nWhile loop example:")
count = 1
while count <= 5:
    print(f"Count: {count}")
    count += 1
```
