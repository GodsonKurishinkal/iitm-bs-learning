## 1. Discrete Random Variables
### 1.1 Theory
A random variable assigns numbers to outcomes. Discrete random variables take countable values.

### 1.2 Mathematical Definition
**PMF (Probability Mass Function)**: $p(x) = P(X = x)$

**Properties**: $\sum_x p(x) = 1$, $p(x) \geq 0$

**CDF**: $F(x) = P(X \leq x) = \sum_{t \leq x} p(t)$

## 2. Continuous Random Variables
### 2.1 Theory
Continuous random variables take any value in an interval. Probability is found via integration.

### 2.2 Mathematical Definition
**PDF (Probability Density Function)**: $f(x)$ where $P(a < X < b) = \int_a^b f(x) dx$

**Properties**: $\int_{-\infty}^{\infty} f(x) dx = 1$, $f(x) \geq 0$

**CDF**: $F(x) = P(X \leq x) = \int_{-\infty}^{x} f(t) dt$
