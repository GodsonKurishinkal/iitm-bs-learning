### 1.1 Theory
Association measures whether the distribution of one variable differs across levels of another. Contingency tables and chi-square tests assess this.

### 1.2 Mathematical Definition
**Chi-square statistic**: $\chi^2 = \sum \frac{(O - E)^2}{E}$

where $O$ = observed, $E$ = expected under independence

## 2. Correlation for Numerical Variables
### 2.1 Theory
Correlation measures linear relationship strength and direction. Pearson's $r$ is most common; Spearman's $\rho$ is robust to outliers and non-linearity.

### 2.2 Mathematical Definition
**Pearson Correlation**:
$$r = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum (x_i - \bar{x})^2 \sum (y_i - \bar{y})^2}}$$

**Properties**: $-1 \leq r \leq 1$; $|r| = 1$ means perfect linear relationship

## 3. Correlation â‰  Causation
### 3.1 Theory
Correlation indicates association, not causation. Confounding variables, reverse causation, or coincidence can create spurious correlations.

### 3.2 Critical Thinking
- **Confounding**: Ice cream sales and drownings (both caused by summer)
- **Reverse causation**: Does advertising increase sales, or do growing sales lead to more ad budget?
- **Spurious**: Nicolas Cage movies and pool drownings
