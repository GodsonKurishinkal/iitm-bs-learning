### 1.2 Mathematical Definition
$$P(n, r) = \frac{n!}{(n-r)!}$$

**Interpretation**: Choose $r$ items from $n$ items where order matters

## 2. Combinations
### 2.1 Theory
Combinations count selections where ORDER DOES NOT MATTER. Selecting 3 items from 5 gives the same combination regardless of order.

### 2.2 Mathematical Definition
$$C(n, r) = \binom{n}{r} = \frac{n!}{r!(n-r)!}$$

**Relationship**: $C(n, r) = P(n, r) / r!$
