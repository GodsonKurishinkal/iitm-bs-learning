## 2. Probability Axioms and Rules
### 2.1 Theory
Probability follows three axioms from which all probability rules can be derived.

### 2.2 Mathematical Definition
**Axioms**:
1. $P(A) \geq 0$
2. $P(S) = 1$
3. For mutually exclusive events: $P(A \cup B) = P(A) + P(B)$

**Derived Rules**:
- Complement: $P(A^c) = 1 - P(A)$
- General Addition: $P(A \cup B) = P(A) + P(B) - P(A \cap B)$
